```python
import matplotlib.pyplot as plt

# Define the model names, FLOPs and inference times
models = ["UNet", "UNet++", "Atten_Unet", "TransCNN"]
flops = [96.682, 238.52, 371.68, 2.4614]  # in GigaFLOPs
inference_times = [0.6852124836001167, 1.6670223311000882, 2.679682940600105, 0.3848439372999565]  # in seconds

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(flops, inference_times, color='blue')
plt.title('FLOPs vs Inference Time')
plt.xlabel('FLOPs (GigaFLOPs)')
plt.ylabel('Inference Time (seconds)')

# Annotate the points with the model names
for i, model in enumerate(models):
    plt.annotate(model, (flops[i], inference_times[i]))

# Show the plot
plt.show()
```
